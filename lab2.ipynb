```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(14, 10))

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

ax.axhline(y=0, color='black', linewidth=1.5)  #  OX
ax.axvline(x=0, color='black', linewidth=1.5)  #  OY

ax.plot(x, y1, 'b-', label=' 1 (sin(x))', linewidth=2)
ax.plot(x, y2, 'r-', label=' 2 (cos(x))', linewidth=2)

x_points = np.array([2, 4, 6, 8])
y_points = np.sin(x_points)
ax.scatter(x_points, y_points, color='green', s=80, zorder=5, label='')

major_xticks = np.arange(0, 11, 2)
major_yticks = np.arange(-1, 1.5, 0.5)
ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)

minor_xticks = np.arange(0, 11, 1)
minor_yticks = np.arange(-1, 1.5, 0.25)
ax.set_xticks(minor_xticks, minor=True)
ax.set_yticks(minor_yticks, minor=True)

ax.tick_params(axis='both', which='major', labelsize=10)

for tick in minor_xticks:
    if tick not in major_xticks:
        ax.text(tick, -0.05, f'{tick}', ha='center', va='top',
                fontsize=8, color='blue', alpha=0.7)

for tick in minor_yticks:
    if tick not in major_yticks:
        ax.text(-0.05, tick, f'{tick:.2f}', ha='right', va='center',
                fontsize=8, color='blue', alpha=0.7)

ax.set_xlabel(' OX', fontsize=12, color='blue', fontweight='bold')
ax.set_ylabel(' OY', fontsize=12, color='blue', fontweight='bold')

ax.set_title('  ', fontsize=16, pad=20, fontweight='bold')

ax.legend(loc='upper right', fontsize=11, framealpha=0.9)

ax.grid(True, which='major', linestyle='-', alpha=0.7, linewidth=1)
ax.grid(True, which='minor', linestyle=':', alpha=0.4, linewidth=0.5)

ax.set_xlim(-0.5, 10.5)
ax.set_ylim(-1.2, 1.2)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(10, 6))

x3 = np.linspace(-3, 4, 400)
y3 = x3**2 - x3 - 6

ax.plot(x3, y3, 'b-', linewidth=2, label='$y(x) = x^2 - x - 6$')
ax.axhline(y=0, color='black', linewidth=0.5, linestyle='--')
ax.axvline(x=0, color='black', linewidth=0.5, linestyle='--')
ax.grid(True, alpha=0.3)
ax.set_xlabel('x')
ax.set_ylabel('y(x)')
ax.set_title('y(x) = x^2 - x - 6')

root1 = -2
root2 = 3
ax.plot(root1, 0, 'ro', markersize=8, label=f': x = {root1}')
ax.plot(root2, 0, 'go', markersize=8, label=f': x = {root2}')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = []

for xi in x:
    inner = 1 / (1 + np.sin(xi) ** 2)

    base = 1 + np.tan(inner)

    log_arg = xi ** 2 + 1

    log_part = np.log(log_arg) / np.log(base)

    exp_part = np.exp(-abs(xi) / 10)

    y.append(log_part * exp_part)


plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.axhline(y=0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(x=0, color='black', linewidth=0.5, linestyle='--')
plt.grid(True, alpha=0.3)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.title('  ')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline

# 1.  
x_edges = [1, 2, 3.7, 4.5, 5.7, 6.4]
heights = [0.28, 0.03, 0.18, 0.26, 0.08]

plt.figure(figsize=(8, 5))

plt.bar(x_edges[:-1], heights, 
        width=np.diff(x_edges), 
        align='edge', 
        color='pink', 
        edgecolor='none')

centers = [(x_edges[i] + x_edges[i+1])/2 for i in range(5)]
x_points = [0.5] + centers + [6.5]
y_points = [0] + heights + [0]

x_smooth = np.linspace(0.5, 6.5, 500)
cs = CubicSpline(x_points, y_points, bc_type='natural')
y_smooth = cs(x_smooth)

plt.plot(x_smooth, y_smooth, color='red', linewidth=2.5)

plt.xlabel('petal_length')
plt.ylabel('Frequency')
plt.xlim(0, 7)
plt.ylim(0, 0.3)
plt.xticks(range(1, 8))
plt.grid(axis='y', alpha=0.3)

plt.show()



#2

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 6))

x = np.linspace(0, 200, 100)
y = x/2

ax.plot(x, y, 'b-', linewidth=2)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title(' y = x')
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')
ax.set_xlim(0, 200)
ax.set_ylim(0, 100)

from mpl_toolkits.axes_grid1.inset_locator import inset_axes

ax_inset = inset_axes(ax, width="30%", height="30%", 
                      loc='right')

ax_inset.plot(x, y, 'r-', linewidth=1.5, color='blue')
ax_inset.set_xlim(0, 200)
ax_inset.set_ylim(0, 100)
ax_inset.set_title('  ')
ax_inset.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()


#3

import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# 1.   y = x/2  100  200
x1 = np.linspace(0, 200, 100)
y1 = x1 / 2

ax1.plot(x1, y1, 'b-', linewidth=2)
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title(' : y = x/2')
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 200)
ax1.set_ylim(0, 100)

# 2.     (0,0)  (100,10000)
x2 = np.linspace(0, 100, 100)
y2 = x2 ** 2

ax2.plot(x2, y2, 'r--', linewidth=2)
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('  : y = x')
ax2.grid(True, alpha=0.3)
ax2.set_xlim(-5, 105)
ax2.set_ylim(-200, 10200)


plt.tight_layout()
plt.show()
```
